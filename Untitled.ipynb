````python
def plot_sound(samples_per_sec, data):
    n_samples = data.shape[0]
    n_channels = data.shape[1]
    duration_sec = n_samples / samples_per_sec
    if n_samples > 1_000_000:
        data = data[::10]
        n_samples = data.shape[0]
    time = np.linspace(0.0, duration_sec, n_samples)
    plt.plot(time, data[:, 0], label="Left channel")
    plt.plot(time, data[:, 1], label="Right channel")
    plt.legend()
    plt.xlabel("Time [s]")
    plt.ylabel("Amplitude")
    plt.show()
````

````python
rate, gong = wav.read("./www/gong-255733.wav")
print(f"Samples per second: {rate}")
print(f"Number of channels: {gong.shape[1]}")
print(f"Data type: {gong.dtype}")
````

````python
plot_sound(rate, gong)
````

````python
duration_sec = 3
n_samples = rate * duration_sec
silence = np.zeros((n_samples, 2), dtype=np.int16)
data = np.vstack((gong, silence))
print(data.shape, data.dtype)
````

````python
plot_sound(rate, data)
````

````python
wav.write("./temp2.wav", rate, data)
````

````python
def create_15_min_chunk(samples_per_sec, gong):
    duration_secs = 15 * 60
    gong_duration_secs = int(gong.shape[0] / samples_per_sec)
    silence_duration_secs = duration_secs - gong_duration_secs
    silence_n_samples = silence_duration_secs * samples_per_sec
    silence = np.zeros((silence_n_samples, 2), dtype=np.int16)
    data = np.vstack((gong, silence))
    return data
````

````python
blocks = [create_15_min_chunk(rate, gong) for _ in range(4)]
type(blocks)
````

````python
data = np.vstack(blocks)
data.shape
````

````python
wav.write("gong_one_hour.wav", rate, data)
````
